```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# create a list to store urls of multiple images
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/888/571"]

# read and display the image
# loops over each image url, multiple images can be stored in the list
for url in urls:
  image = io.imread(url) # read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) # resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB) # convert color to RGB
  final_frame = cv.hconcat((image, image_2)) # concatenate image
  cv2_imshow(final_frame) # show image
  print('\n')
```

##5. Explain the relationship between pixels and high or low image resolution!

####**Answer :**

Image resolution refers to the number of pixels present in an image and impacts how detailed and sharp the image appears. Pixel is the smallest unit of a digital image that represents a single color point in an image.

Relationship:
- Low resolution images have fewer pixels per unit area. Makes images appear coarser and less detailed. When zoomed in, the pixels are more visible ( "pixelated" effect)
- High resolution images have more pixels per unit area. Makes images look more detailed and sharp

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# draw two diagonal lines (cross) from corner to corner
for i in range(min(height, width)):
    image_3[i, i] = [255, 255, 255]
    image_3[height - i - 1, i] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

line_length = 120
middle_y = 125

# Set the positions
start_x = 125
end_x = start_x + line_length

# Draw a short horizontal line
image_3[middle_y, start_x:end_x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
